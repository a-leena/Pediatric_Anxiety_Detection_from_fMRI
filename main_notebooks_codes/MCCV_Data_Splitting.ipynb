```python
import random
import os
from glob import glob
import json
```

```python
data_path = os.path.join('..', 'Pediatric_Anxiety_Disorder', 'sub-*')
```

```python
subjects = [int(os.path.basename(f).split('-')[1]) for f in glob(data_path)]
print(len(subjects))
print(subjects)
```

```python
def create_repeated_holdouts(sub_ids, n_holdouts=5, slglm_size=60, start_seed=0):    
    holdouts = []
    seen_slglm_splits = set()
    current_seed = start_seed
    
    while len(holdouts) < n_holdouts:
        # setting seed (used for reproducibility)
        random.seed(current_seed)
        
        # randomly picking `slglm_size` subjects
        slglm_split = random.sample(sub_ids, slglm_size)

        # checking and adding unique splits only
        split_tuple = tuple(sorted(slglm_split))
        if split_tuple not in seen_slglm_splits:
            seen_slglm_splits.add(split_tuple)

            # adding remaining subjects to ml_split
            ml_split = [sub for sub in sub_ids if sub not in slglm_split]

            # creating holdout
            holdouts.append({
                'holdout_id': len(holdouts)+1,
                'seed': current_seed,
                'slglm_subjects': slglm_split,
                'ml_subjects': ml_split
            })
            
        current_seed += 1
        
    return holdouts
```

## Generating 30 unique holdouts

```python
holdouts30 = create_repeated_holdouts(subjects, n_holdouts=30)
```

```python
for holdout in holdouts30:
    for k,v in holdout.items():
        print(f"{k}: {v}\n")
    print()
```

```python
holdouts30_dict = {}
for holdout in holdouts30:
    holdouts30_dict[f"holdout-{holdout['holdout_id']:02d}"] = {
        'seed': holdout['seed'],
        'slglm': holdout['slglm_subjects'],
        'ml': holdout['ml_subjects']
    }
print(holdouts30_dict)
```

```python
path = os.path.join('..', 'dataset_holdouts', '30_holdouts_new2')
os.makedirs(path, exist_ok=True)
filepath = os.path.join(path, 'holdouts30_split_ids.json')
with open(filepath, 'w') as json_file:
    json.dump(holdouts30_dict, json_file, indent=4)
```
