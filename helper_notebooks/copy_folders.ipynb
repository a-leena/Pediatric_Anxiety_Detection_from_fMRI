```python
ROOT_FOLDER = os.path.join('..', 'dataset_holdouts', '10_holdouts')
CSV_FILENAME = 'ML_Dataset_holdout-*.csv'
```

```python
OLD_COL_NAME = 'subject_id'
NEW_COL_NAME = 'participant_id'
```

```python
def modify_value(val):
    return 'sub-0'+str(val)
```

```python
def modify_csv(filepath):
    df = pd.read_csv(filepath)
    if OLD_COL_NAME not in df.columns:
        print(f"[SKIP] Column '{OLD_COL_NAME}' not found in: {filepath}")
        return

    df.rename(columns={OLD_COL_NAME: NEW_COL_NAME}, inplace=True)
    df[NEW_COL_NAME] = df[NEW_COL_NAME].apply(modify_value)
    df.to_csv(filepath, index=False)
    print(f"[OK] Processed: {filepath}")
```

```python
pattern = os.path.join(ROOT_FOLDER, '**', CSV_FILENAME)
csv_files = glob.glob(pattern, recursive=True)
csv_files
```

```python
print(f"Found {len(csv_files)} CSV file(s). Processing...\n")
for filepath in csv_files:
    modify_csv(filepath)
print("\nDone!")
```
